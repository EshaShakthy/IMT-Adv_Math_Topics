```
Since the duration of infectiousness is D, what is the probability that an infected person
becomes removed during this time step?
```

$3=\dfrac {1} {D}$

$4=\dfrac {I} {D}$

```
import matplotlib.pyplot as plt
```

```

def SIR_Model (runtime , D , R_0):
  S=99999
  I=1
  R=0
  t=1
  S_list=[S]
  I_list=[I]
  R_list=[R]
  t_list=[1]
  for t in range (1,runtime):
    new_infected=R_0/D*I
    recoved_new=I/D 
    S-=new_infected
    I+=new_infected-recoved_new
    R+=recoved_new
    S_list.append(S)
    I_list.append(I)
    R_list.append(R)
    t_list.append(t)
  plt.plot(t_list,S_list)
  plt.plot(t_list,R_list)
  plt.plot(t_list,I_list)
  plt.show()  
```

Code the model that represents this situation. Produce graphs of the susceptible, infected,
and removed populations over time. Start with D being 5 days, and R0 = 3.
