```
import random
import matplotlib.pyplot as plt
#Model 1
def run_model(steps):
  new_infected_numbers = [1]
  total_infected_numbers  = [1]
  for i in range(steps):
    infected_people_now = new_infected_numbers[-1]
    infected_people_next = 0
    for j in range(infected_people_now):
      infected_people_next += random.randint(0,2)
    new_infected_numbers.append(infected_people_next)
    total_infected = total_infected_numbers[-1] + infected_people_next
    total_infected_numbers.append(total_infected)
  print(total_infected_numbers)
  return total_infected_numbers

trajectories = []
for i in range(10):
  trajectories.append(run_model(20))


times = []
for t in range(21):
  times.append(t)

fig = plt.figure()
for trajectory in trajectories:
  plt.plot(times,trajectory)
plt.xlabel("Time")
plt.ylabel("Total Infections")
plt.show()






```

```
import random
import matplotlib.pyplot as plt
#Model 2
def run_model(steps,prob0,prob1,prob2):
 new_infected_numbers = [1]
 total_infected_numbers  = [1]
 for i in range(steps):
   infected_people_now = new_infected_numbers[-1]
   infected_people_next = 0
   for j in range(infected_people_now):
     rv = random.random()
     if(rv<=prob0):
       new = 0
     else:
       if(rv<=(prob0+prob1)):
         new = 1
       else:
         new = 2
     infected_people_next += new
   new_infected_numbers.append(infected_people_next)
   total_infected = total_infected_numbers[-1] + infected_people_next
   total_infected_numbers.append(total_infected)
 #print(total_infected_numbers)
 return total_infected_numbers
 
trajectories = []
for i in range(50):
 trajectories.append(run_model(20,0.2,0.25,0.55))
 
 
times = []
for t in range(21):
 times.append(t)
 
fig = plt.figure()
for trajectory in trajectories:
 plt.plot(times,trajectory)
plt.xlabel("Time")
plt.ylabel("Total Infections")
plt.title("Total Infections Over Time")
plt.show()

```
